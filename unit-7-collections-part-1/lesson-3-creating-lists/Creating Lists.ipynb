# Creating Lists

### Literally

```python
emails = ['john@gmail.com', 'jane@live.com', 'mary@rmotr.com']
print(emails)
```

### Programmatically

```python
emails = []
emails.append('john@gmail.com')
emails.append('jane@live.com')
emails.append('mary@rmotr.com')
print(emails)
```

### Programmatically vs Literally

```python
emails = ['john@gmail.com', 'jane@live.com']
emails.append('mary@rmotr.com')
print(emails)
```
